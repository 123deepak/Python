```python
#If you read the docs for OneHotEncoder you'll see the input for fit is "Input array of type int". 
#So you need to do two steps for your one hot encoded data
import pandas as pd
from collections import defaultdict
from sklearn.preprocessing import OneHotEncoder
result = pd.DataFrame({
            "X": ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm'],
            "Y": ['what', 'this', 'is', 'what', 'is', 'what', 'this', ]})


d = defaultdict(LabelEncoder)
result = result.apply(lambda x: d[x.name].fit_transform(x))
result
```

```python
OHE = OneHotEncoder(handle_unknown='ignore')
result = OHE.fit_transform(result).toarray()
result
```
